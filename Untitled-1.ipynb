```python


url = 'https://ptmi2.intel.com/P1276/scorecard.php'
response = requests.get(url, auth=HTTPKerberosAuth(mutual_authentication=OPTIONAL), verify=False)
html_content = response.text
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
table = soup.find('table', {'class': 'CSIDataTable'})
```

```python
df = pd.read_html(str(table))[0]
df.columns = df.iloc[0]
df = df.drop(df.index[0])
df.columns = df.columns.str.replace('sort$', '', regex=True)
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df = df[df['FAB']!='D1D']
df.reset_index(inplace=True, drop=True)

# Create a list of the column names with 'A' and 'B' at the beginning
cols = ['FAB', 'PRODUCT NAME'] + [col for col in df.columns if col not in ['FAB', 'PRODUCT NAME']]

# Reorder the DataFrame columns using the new list
df = df[cols]
df = df.sort_values(by='PRODUCT NAME', ascending=True)

for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')

df.dtypes
```

```python
# Initialize the dictionary to store the column mappings
column_mappings = {}

# Define the patterns to match
patterns = ['CAP', 'IDV', 'SICC']

# Loop through each pattern and find matching columns
for pattern in patterns:
    # Find columns that contain the pattern followed by "TARGET" or "COMMIT"
    target_commit_columns = df.columns[df.columns.str.contains(f'^{pattern}.* (TARGET|COMMIT)')]

    # Find the base columns by excluding the "TARGET" or "COMMIT" columns
    # This assumes that the base column starts with the pattern and does not contain "TARGET" or "COMMIT"
    base_columns = df.columns[(df.columns.str.startswith(pattern)) & (~df.columns.isin(target_commit_columns))]

    # Assign the found column names to the dictionary
    # If there are multiple matches, you can handle them as needed (e.g., take the first one)
    if not base_columns.empty:
        column_mappings[pattern] = base_columns[0]
    if not target_commit_columns.empty:
        # Since there will only be either "TARGET" or "COMMIT", we can safely take the first match
        column_mappings[f'{pattern} TARGET'] = target_commit_columns[0]

# Now `column_mappings` contains the mappings
column_mappings
```

```python
excel_filename = 'output.xlsx'
```

```python
def autofit_sheet(df,ws):
    for i,col in enumerate(df.columns):
        column_len = df[col].astype(str).map(len).max()
        column_len = max(column_len,len(col))+2
        ws.set_column(i,i,column_len)
```

```python
with pd.ExcelWriter(excel_filename, engine='xlsxwriter') as writer:
    # Write the DataFrame to the Excel file
    df.to_excel(writer, index=False, sheet_name='Sheet1')

    # Access the XlsxWriter workbook and worksheet objects
    workbook = writer.book
    worksheet = writer.sheets['Sheet1']

    # Define a format for the cells that meet the condition
    red_format = workbook.add_format({'bg_color': '#FFC7CE', 'font_color': '#9C0006'})
    orange_format = workbook.add_format({'bg_color': '#FFEB9C', 'font_color': '#9C6500'})
    green_format = workbook.add_format({'bg_color': '#C6EFCE', 'font_color': '#006100'})

    # Get the number of rows and columns in the DataFrame
    (max_row, max_col) = df.shape

    param_list = ['CAP','IDV','SICC']

    for param in param_list:
        
        base_col = column_mappings[f'{param}']
        target_col = column_mappings[f'{param} TARGET']

        print(f'{base_col}, {target_col}')
        
        base_excel_col = xlsxwriter.utility.xl_col_to_name(df.columns.get_loc(base_col))
        target_excel_col = xlsxwriter.utility.xl_col_to_name(df.columns.get_loc(target_col))

        # Define the conditional formatting rule
        # Note: Excel row indexing starts from 1, and the header is row 1
        for row in range(2, max_row + 2):  # Start from row 2 to skip the header
            
            if param in ['CAP','IDV']:

                worksheet.conditional_format(f'{base_excel_col}{row}',
                                                {'type': 'formula',
                                                'criteria': f'={base_excel_col}{row}<{target_excel_col}{row} * 0.99',
                                                'format': red_format})
                worksheet.conditional_format(f'{base_excel_col}{row}',
                                             {'type': 'formula',
                                              'criteria': f'={base_excel_col}{row} < {target_excel_col}{row} * 0.995',
                                              'format': orange_format})
                # # Green: Otherwise
                worksheet.conditional_format(f'{base_excel_col}{row}',
                                             {'type': 'formula',
                                              'criteria': f'={base_excel_col}{row} >= {target_excel_col}{row} * 0.995',
                                              'format': green_format})
            elif param == 'SICC':
                # Red: Base value is more than 20% above target
                worksheet.conditional_format(f'{base_excel_col}{row}',
                                             {'type': 'formula',
                                              'criteria': f'={base_excel_col}{row} > {target_excel_col}{row} * 1.2',
                                              'format': red_format})
                # # Orange: Base value is 10-20% above target
                worksheet.conditional_format(f'{base_excel_col}{row}',
                                             {'type': 'formula',
                                              'criteria': f'={base_excel_col}{row} > {target_excel_col}{row} * 1.1',
                                              'format': orange_format})
                # # Green: Otherwise
                worksheet.conditional_format(f'{base_excel_col}{row}',
                                             {'type': 'formula',
                                              'criteria': f'={base_excel_col}{row} <= {target_excel_col}{row} * 1.1',
                                              'format': green_format})
    autofit_sheet(df,worksheet)
```
